## import library

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.animation as animation
%matplotlib inline
sns.set();
```

## 関数描画

```python
# 0 < x < 1
x = np.arange(0, 1, 0.00001)

# y = x^2 + 10x + 1*/
y = 6*x*(1-x)

# グラフへのプロット実行
plt.plot(x, y)
plt.fill(x,y,color="b",alpha=0.5)
plt.show()
```
